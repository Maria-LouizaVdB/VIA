# Grid world

Suppose you are standing on a 2x2 square. If you are standing on the bottom right square (br), the game ends. As you move around the square, you win rewards. You want to find the way to win the most rewards (i.e. find the optimum path to end the game). 

To simplify the set up, we assume you can move up (u), down (d), left (l) or right (r) from any square. However, the probability that you move outside the square is $0$. So, the only non-zero probabilities are:

| State (s) | Action (a) | Next State (s') | Probability |
|-----------|------------|----------------|------------|
| TL        | R          | TR             | 0.9        |
| TL        | R          | BL             | 0.1        |
| TL        | D          | BL             | 0.9        |
| TL        | D          | TR             | 0.1        |
| TR        | L          | TL             | 0.9        |
| TR        | L          | BR             | 0.1        |
| TR        | D          | BR             | 0.8        |
| TR        | D          | TL             | 0.2        |
| BL        | R          | BR             | 0.9        |
| BL        | R          | TL             | 0.1        |
| BL        | U          | TL             | 0.8        |
| BL        | U          | BR             | 0.2        |
| BR        | U          | BR             | 1.0        |
| BR        | D          | BR             | 1.0        |
| BR        | R          | BR             | 1.0        |
| BR        | L          | BR             | 1.0        |

(To incooperate the terminating bottom right square, we set the transition probability from bottom right to bottom right to $1$ always.)

Similarly, the only non-zero rewards are:

| State (s) | Action (a) | Next State (s') | Reward |
|-----------|------------|----------------|--------|
| TL        | R          | TR             | -1.0   |
| TL        | R          | BL             | -2.0   |
| TL        | D          | BL             | -2.0   |
| TL        | D          | TR             | -1.0   |
| TR        | L          | TL             | -1.5   |
| TR        | L          | BR             | 10.0   |
| TR        | D          | BR             | 15.0   |
| TR        | D          | TL             | -1.0   |
| BL        | R          | BR             | 20.0   |
| BL        | R          | TL             | -2.5   |
| BL        | U          | TL             | -0.5   |
| BL        | U          | BR             | 5.0    |

## Creating the transition and reward functions

When setting up the problem, we assume it is possible to choose to move in any direction but the probability of this happening is $0$. So we create an initial transition dictionary where all possible transitions have value $0$ and then update this to reflect the non-zero probabilities. This dictionary is then stored inside a function, `prob_grid_world`, so that it can be passed into the value iteration algorithm function, `value_iteration_function`.

## Implementing value iteration algorithm

```python
from value_iteration import value_iteration_function, prob_grid_world, reward_grid_world

v,pi,k = value_iteration_function(states = ["tl","tr","bl","br"],
                                  actions = ["r","l","u","d"],
                                  trans_fun = prob_grid_world,
                                  reward_fun = reward_grid_world,
                                  epsilon = 0.05,
                                  gamma = 0.9,
                                  max_iteration = 100)

print("Value function:", v)
print("Optimal policy:", pi)
print("It took",k,"iterations to get these results.")
```

The optimum policy if you are in the:
* top left square is to move downwards,
* top right square is to move downwards,
* and bottom left square is to move right.

## Why does the bottom right have an optimum move if it terminates?

The output says that the optimum move from bottom right is to move right, even though it is terminal. This stems comes from our implementation that doesn't truly include the terminating step. Instead of fully terminating, we make the bottom right square recurrent with $0$ future rewards. Thus, the algorithm tries to find the optimal action. 

The action is chosen by picking the action with the greatest Q-value, which is calculated using numpy.argmax on the list $[0,0,0,0]$. This will always return the index $0$, which in this case corresponsd to action "r". As seen in the following code example, if the action set was coded `["l","u","d","r"]`, it would return "l". 

```python
v,pi,k = value_iteration_function(states = ["tl","tr","bl","br"],
                                  actions = ["l","u","d","r"],
                                  trans_fun = prob_grid_world,
                                  reward_fun = reward_grid_world,
                                  epsilon = 0.05,
                                  gamma = 0.9,
                                  max_iteration = 100)

print("Value function:", v)
print("Optimal policy:", pi)
print("It took",k,"iterations to get these results.")
```
