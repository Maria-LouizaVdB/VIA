# Should you party or rest when you are healthy or sick?

This example exercise stems from Chapter 9, Exercise 27 in Artificial Integlligence: Foundations of Computational Agents by David L. Poole and Alan K. Mackworth (available at https://artint.info/2e/html2e/ArtInt2e.Ch9.S5.html). 

Suppose Sam really likes to party but is worried about being sick. He wants to decide what his optimal action should be. He is either healthy or sick and can either party or rest. Thus, the state space is $S=\{\text{healthy},\text{sick}\}$ and the action space is $A=\{\text{party},\text{rest}\}$.

He estimates that the probability of his being healthy tomorrow, given his current state and his action is:

| State   | Action | Next State | Probability |
| ------- | ------ | ---------- | ----------- |
| healthy | relax  | healthy    | 0.95        |
| healthy | relax  | sick       | 0.05        |
| healthy | party  | healthy    | 0.70        |
| healthy | party  | sick       | 0.30        |
| sick    | relax  | healthy    | 0.50        |
| sick    | relax  | sick       | 0.50        |
| sick    | party  | healthy    | 0.10        |
| sick    | party  | sick       | 0.90        |

He prefers partying over relaxing. However, he doesn't like being sick. The reward of each state and action is:
| State   | Action | Reward |
| ------- | ------ | ------ |
| healthy | relax  | 7      |
| healthy | party  | 10     |
| sick    | relax  | 0      |
| sick    | party  | 2      |

He discounts previous experience with a discount factor $\gamma = 0.9$. 

## Optimal action using value iteration

We suppose the initial value of each state is $0$, $$V(\text{healthy}) = 0, V(\text{sick}) = 0.$$ We consider how these values would change after one action for healthy:
$$Q(\text{healthy},\text{relax}) = \text{Reward}(\text{healthy},\text{relax}) + \gamma \times \left( P(\text{healthy}|\text{healthy},\text{relax})V(\text{healthy}) + P(\text{sick}|\text{healthy},\text{relax})V(\text{sick})\right)$$ 
$$= 7+0.9Ôêù(0.95Ôêù0+0.05Ôêù0)=7,$$

$$Q(\text{healthy},\text{party}) = \text{Reward}(\text{healthy},\text{party}) + \gamma \times \left( P(\text{healthy}|\text{healthy},\text{party})V(\text{healthy}) + P(\text{sick}|\text{healthy},\text{party})V(\text{sick})\right)$$ 
$$=10+0.9Ôêù(0.7Ôêù0+0.3Ôêù0)=10,$$

$$V(\text{healthy}) = \max\{ Q(\text{healthy},\text{relax}), Q(\text{healthy},\text{party}) \} = 10.$$

After one action for sick:
$$Q(\text{sick},\text{relax}) = \text{Reward}(\text{sick},\text{relax}) + \gamma \times \left( P(\text{healthy}|\text{sick},\text{relax})V(\text{healthy}) + P(\text{sick}|\text{sick},\text{relax})V(\text{sick})\right)$$ 
$$= 0+0.9Ôêù(0.5Ôêù0+0.5Ôêù0)=0,$$

$$Q(\text{sick},\text{party}) = \text{Reward}(\text{sick},\text{party}) + \gamma \times \left( P(\text{healthy}|\text{sick},\text{party})V(\text{healthy}) + P(\text{sick}|\text{sick},\text{party})V(\text{sick})\right)$$ 
$$=2+0.9Ôêù(0.1Ôêù0+0.9Ôêù0)=2,$$

$$V(\text{sick}) = \max\{ Q(\text{sick},\text{relax}), Q(\text{sick},\text{party}) \} = 2.$$

The value functions are now:
$$V(\text{healthy}) = 10, V(\text{sick}) = 2.$$

### Value iteration

We perform another round of updates:
$$Q(\text{healthy},\text{relax}) = 7+0.9Ôêù(0.95Ôêù10+0.05Ôêù2)= 15.64,$$
$$Q(\text{healthy},\text{party}) = 10+0.9Ôêù(0.7Ôêù10+0.3Ôêù2)=16.84,$$
$$Q(\text{sick},\text{relax}) = 0+0.9Ôêù(0.5Ôêù16.84+0.5Ôêù2)=8.478,$$
$$Q(\text{sick},\text{party}) = 2+0.9Ôêù(0.1Ôêù16.84+0.9Ôêù2)=5.1356,$$ 
and thus 
$$V(\text{healthy}) = 16.84, V(\text{sick}) = 5.1356.$$ 

We perform a final third round of updates:
$$Q(\text{healthy},\text{relax}) = 7+0.9Ôêù(0.95Ôêù16.84+0.05Ôêù8.478)\approx 21.78$$
$$Q(\text{healthy},\text{party}) = 10+0.9Ôêù(0.7Ôêù16.84+0.3Ôêù8.478) \approx 22.90,$$
$$Q(\text{sick},\text{relax}) = 0+0.9Ôêù(0.5Ôêù22.898+0.5Ôêù8.478) \approx 14.12,$$
$$Q(\text{sick},\text{party}) = 2+0.9Ôêù(0.1Ôêù22.898+0.9Ôêù8.478) \approx 10.93,$$ 
and thus 
$$V(\text{healthy}) = 22.90, V(\text{sick}) = 14.12.$$ 

### Policy

When Sam is healthy, he should party as $Q(\text{healthy},\text{party}) > Q(\text{healthy},\text{relax})$. When Sam is sick, he should relax as $Q(\text{healthy},\text{party}) > Q(\text{healthy},\text{relax})$.

## Using code

Before we implement value iteration, we convert the transition and reward tables into functions that use dictionaries. Respectively, the functions are `prob_27` and `reward_27`. These are in the file titled `happy_sick_functions.py`

```python
from happy_sick_functions import prob_27, reward_27
```

```python
from via_function import value_iteration_function
```

```python
v,pi,k = value_iteration_function(states = ["h","s"],
                                actions = ["p","r"],
                                trans_fun = prob_27,
                                reward_fun = reward_27,
                                epsilon = 0.05,
                                gamma = 0.9,
                                max_iteration = 2)

print("Value function:", v)
print("Optimal policy:", pi)
print("The algorithm performed",k,"iterations to get these results.")
```

Thus, we get the same optimal policy as we calculated manually: when Sam is healthy, he should party and when Sam is sick, he should relax. The value function values are also similar to the previously manually calculated values.

### Until convergence
We can continue until this is stable and converges by increasing the `max_iteration` argument.

```python
v,pi,k = value_iteration_function(states = ["h","s"],
                                actions = ["p","r"],
                                trans_fun = prob_27,
                                reward_fun = reward_27,
                                epsilon = 0.05,
                                gamma = 0.9,
                                max_iteration = 50)

print("Value function:", v)
print("Optimal policy:", pi)
print("The algorithm performed",k,"iterations to get these results.")
```

We obain the same optimal policy but the value function has increased.

#### $\epsilon$ criteria
We can also change the convergence criteria is we want to decrease the number of required iterations to converge. We do this by changing the `epsilon` argument.

```python
v,pi,k = value_iteration_function(states = ["h","s"],
                                actions = ["p","r"],
                                trans_fun = prob_27,
                                reward_fun = reward_27,
                                epsilon = 0.1,
                                gamma = 0.9,
                                max_iteration = 50)

print("Value function:", v)
print("Optimal policy:", pi)
print("The algorithm performed",k,"iterations to get these results.")
```

The optimal policy is the same. We converse $6$ iterations earlier and thus have lower value function values.
